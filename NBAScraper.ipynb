```python
import csv
import requests
from bs4 import BeautifulSoup

# URL of the website to scrape
url = "https://www.espn.com/nba/boxscore/_/gameId/401547678"

# Send a GET request to the website
response = requests.get(url)

# Create a BeautifulSoup object with the website's HTML content
soup = BeautifulSoup(response.content, "html.parser")
print(soup.find_all("table"))
```

```python

# Find the "PageLayout_Main" div on the page
page_layout_main = soup.find("div", class_="PageLayout__Main")

if page_layout_main:
    # Find all tables within the "PageLayout_Main" div
    tables = page_layout_main.find_all("table")

    # Specify the output CSV file path
    csv_file = "output.csv"

    with open(csv_file, "w", newline="") as file:
        writer = csv.writer(file)

        for i in range(0, len(tables), 2):
            table1 = tables[i]
            table2 = tables[i+1] if i+1 < len(tables) else None

            # Extract the table headers
            headers1 = [header.text for header in table1.find_all("th")]
            headers2 = [header.text for header in table2.find_all("th")] if table2 else []

            # Extract the table rows and their data
            rows1 = table1.find_all("tr")
            data1 = []
            for row in rows1:
                row_data = [cell.text for cell in row.find_all("td")]
                if row_data:
                    data1.append(row_data)

            rows2 = table2.find_all("tr") if table2 else []
            data2 = []
            for row in rows2:
                row_data = [cell.text for cell in row.find_all("td")]
                if row_data:
                    data2.append(row_data)

            # Append the data of the second table to the first table
            appended_data = [data1[i] + data2[i] for i in range(min(len(data1), len(data2)))]

            # Write the combined data into the CSV file
            writer.writerow(headers1 + headers2)  # Write header row
            writer.writerows(appended_data)  # Write the combined data rows

    print(f"Scraped table data successfully saved to '{csv_file}'.")
else:
    print("No 'PageLayout_Main' div found on the webpage.")
```
